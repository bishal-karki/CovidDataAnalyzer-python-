```python
import sqlite3
import pandas as pd
```

```python
connection = sqlite3.connect("covid.db")
query1="SELECT Covid.currentState as Current_State,count(*) as Total FROM 'Covid','District' on Covid.district=District.id where District.title='Kaski' group by currentState"
df1=pd.read_sql_query(query1,connection)

# total_bargraph.savefig('total_covid_data.png')
```

```python
total_bargraph=df1.plot.bar(title='Total covid data Districtwise:',x='Current_State', rot=0)
```

```python
query2="SELECT Covid.currentState as Current_State,count(*) as Total FROM 'Covid','District' on Covid.district=District.id where District.title='Kaski' and lower(Covid.gender)='male' group by Covid.currentState "
df2=pd.read_sql_query(query2,connection)
df2.head
```

```python
total_bargraph=df2.plot.bar(title='Total covid data of male Districtwise:',x='Current_State', rot=0)
```

```python
query3="""  SELECT currentState,SUM(CASE WHEN Covid.age < 18 THEN 1 ELSE 0 END) AS [Under 18],
                SUM(CASE WHEN Covid.age BETWEEN 18 AND 30 THEN 1 ELSE 0 END) AS [18-30],
                SUM(CASE WHEN Covid.age BETWEEN 31 AND 50 THEN 1 ELSE 0 END) AS [31-50],
                SUM(CASE WHEN Covid.age > 50 THEN 1 ELSE 0 END) AS [ Above 50]
                FROM Covid,District on Covid.district=District.id where District.title='Kaski' group by Covid.currentState """
```

```python
df3=pd.read_sql_query(query3,connection)
df3.head
```

```python
total_bargraph_age=df3.plot.bar(title='Total covid data of Age Group Districtwise:',x='currentState', rot=0)
```
